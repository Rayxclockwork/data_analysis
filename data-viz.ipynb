```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
music_df = pd.read_csv('./1000Tracks.csv')
books_df = pd.read_csv('./Goodreads_best1500books.csv')
music = music_df.drop(['album', 'id', 'acousticness', 'danceability', 'duration_ms', 'energy', 'instrumentalness', 'key', 'liveness', 'loudness', 'mode', 'speechiness', 'tempo', 'time_signature', 'valence'], axis=1)
music.head(5)
```

```python
music_df.info()
```

```python
music.describe()
```

```python
sns.set()
plt.figure(figsize=(10, 6), dpi=300)
sns.countplot(x='genre', data=music, palette='Set2')
```

```python
books = books_df.drop(['author_name', 'year_published', 'edition_language', 'avg_rating', 'no_of_raters', 'score', 'no_of_ppl_voted', 'book_url'], axis=1)
books.head(5)
```

```python
books_df.info()
```

```python
books.describe()
```

```python
sns.set()
plt.figure(figsize=(20, 8), dpi=300)
sns.countplot(y='book_genre', data=books, palette='Set3')
```
